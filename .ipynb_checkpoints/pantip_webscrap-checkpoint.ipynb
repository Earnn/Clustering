```python
#https://shopee.co.th/flash_sale/
#https://shopee.co.th/api/v1/category_list/

# read in a page and convert requests text into 'soup' object
r = requests.get('https://shopee.co.th/')
soup = BeautifulSoup(r.text, "html.parser")

#print(soup)

# find the section of relevant links and then parse into iterable rows
#link = 'https://shopee.co.th/' + link_rowname + '-cat.' + link_rowid + '?page=0&sortBy=sales'
#https://shopee.co.th/api/v2/search_items/?by=sales&limit=50&newest=0&order=desc&page_type=search&match_id=9867

category_links = []
links_sections = soup.find_all(name='div', attrs={'class':'category-item clickable_area '})
for links_section in links_sections: 
    link_rowid = links_section['data-category-id']
    link_rowname = links_section['data-catname']      
    link = 'https://shopee.co.th/api/v2/search_items/?by=sales&limit=50&newest=0&order=desc&page_type=search&match_id=' + link_rowid
    #print (link)
    category_links.append(link)
    

# create a list of category links
#category_links = ['https://shopee.co.th/' + row.a['href'] + '-cat.' + for row in link_rows]

# function that takes a link and returns a dictionary of info about that page
def get_category_winners(category_link):
    #print(category_link)
    r = requests.get(category_link, headers={'User-Agent': 'Custom'})    
    soup = BeautifulSoup(r.content, "html.parser")    
    rows = str(soup).split('"itemid"')    
    return rows
    #return {"category":     soup.find(name='title'),
    #        "product":      soup.find(name='div', attrs={'class':'shopee-item-card__text-name'}),
            #"runners_up":   [h2.string for h2 in soup.find_all(name='h2', attrs={'class':'boc2'})]
    #        }

# create list of dictionaries for the first three links
from time import sleep
winners = []
for category_link in category_links[0:1]:
    winners.append(get_category_winners(category_link))
    print('.')
    sleep(1)

# 'pretty print' the winners data
from pprint import pprint
pprint(winners)
```
